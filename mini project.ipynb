```python
# 1. PARENT CLASS

class SmartDevice:
    def __init__(self, device_id: str, name: str):
        if not device_id.strip():
            raise ValueError("Device ID cannot be empty.")

        self.__device_id = device_id
        self.__power_status = False
        self.name = name

    def get_device_id(self) -> str:
        return self.__device_id

    def get_power_status(self) -> bool:
        return self.__power_status

    def turn_on(self):
        self.__power_status = True
        print(f"{self.name} is now turned ON.")

    def turn_off(self):
        self.__power_status = False
        print(f"{self.name} is now turned OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Device ID: {self.__device_id}")
        print(f"Device Name: {self.name}")
        print(f"Power Status: {status}")


# 2. CHILD CLASSES

class SmartThermostat(SmartDevice):
    def __init__(self, device_id: str, name: str, current_temp: float = 22.0):
        super().__init__(device_id, name)
        self.temperature = current_temp

    def read_temperature(self):
        print(f"[{self.name}] Current Temperature: {self.temperature}°C")


class SmartLight(SmartDevice):
    def __init__(self, device_id: str, name: str, brightness: int = 50):
        super().__init__(device_id, name)
        self.__brightness = brightness

    def get_brightness(self) -> int:
        return self.__brightness

    def set_brightness(self, value: int):
        if 0 <= value <= 100:
            self.__brightness = value
            print(f"[{self.name}] Brightness adjusted to {self.__brightness}%.")
        else:
            print("Error: Brightness must be between 0 and 100.")


class SmartCamera(SmartDevice):
    def __init__(self, device_id: str, name: str):
        super().__init__(device_id, name)
        self.is_recording = False

    def start_recording(self):
        if not self.get_power_status():
            print(f"Cannot start recording. {self.name} is currently OFF.")
            return
        self.is_recording = True
        print(f"[{self.name}] Recording started successfully.")

#(Polymorphism)

def run_management_system():
    thermostat = SmartThermostat("T01", "Main Thermostat")
    light = SmartLight("L01", "Living Room Light")
    camera = SmartCamera("C01", "Front Door Camera")

    devices = [thermostat, light, camera]

    while True:
        print("\n" + "="*40)
        print("  SMART DEVICE MANAGEMENT INTERFACE")
        print("="*40)
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature (Thermostat Only)")
        print("5. Adjust Brightness (Light Only)")
        print("6. Start Recording (Camera Only)")
        print("7. Exit")

        choice = input("Select an option (1-7): ").strip()

        if choice == "1":
            print("\n--- Current Device Registry ---")
            for dev in devices:
                dev.display_info()
                print("-" * 20)

        elif choice == "2":
            print("\nWhich device to turn ON?")
            for i, dev in enumerate(devices, 1):
                print(f"{i}. {dev.name}")
            try:
                ch = int(input("Selection: ")) - 1
                if 0 <= ch < len(devices):
                    devices[ch].turn_on()
                else:
                    print("Error: Device selection out of range.")
            except ValueError:
                print("Error: Invalid input. Please enter a valid number.")

        elif choice == "3":
            print("\nWhich device to turn OFF?")
            for i, dev in enumerate(devices, 1):
                print(f"{i}. {dev.name}")
            try:
                ch = int(input("Selection: ")) - 1
                if 0 <= ch < len(devices):
                    devices[ch].turn_off()
                else:
                    print("Error: Device selection out of range.")
            except ValueError:
                print("Error: Invalid input. Please enter a valid number.")

        elif choice == "4":
            thermostat.read_temperature()

        elif choice == "5":
            try:
                val = int(input(f"Enter brightness level for {light.name} (0-100): "))
                light.set_brightness(val)
            except ValueError:
                print("Error: Invalid input! Please enter an integer between 0 and 100.")

        elif choice == "6":
            camera.start_recording()

        elif choice == "7":
            print("Exiting system.")
            break
        else:
            print("Invalid option. Please select a number between 1 and 7.")

if __name__ == "__main__":
    run_management_system()
```
